```python
import pandas as pd
import numpy as np
import datetime as dt
import scipy.special as sps
from scipy import *
import warnings
from scipy.stats import kurtosis
from scipy.stats import skew
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import normalized_mutual_info_score
```

```python
def my_basic_research(df, column):
    print("Базовые метрики")
    print(df[column].describe())
    print("------------------------------------")
    
    print("Самые популярные значения метрики, топ 5")
    print(df[column].value_counts().nlargest(5))
    print("------------------------------------")
    
    print("Эксцесс ", kurtosis(df[column]))
    print("Ассиметрия ", skew(df[column]))
    print("------------------------------------")
    
    
    
    sns.distplot(df[column])
    plt.title("Распределение метрики")
```

# New Billionaire Indicator

```python
missing_values = ["inf","NaN"]
IRI = 'D:\\Michael\\Data\\Ratings\\Billionaire\\Edgar_countBilli.csv'
dat = pd.read_csv(IRI)
dat = dat.drop(columns = ['HDSymbolId', 'Unnamed: 4', 'Unnamed: 5'])
dat = dat.dropna()

Q55 = dat['count_of_billionaires'].quantile(0.55)
Q79 = dat['count_of_billionaires'].quantile(0.79)
Q82 = dat['count_of_billionaires'].quantile(0.825)
Q94 = dat['count_of_billionaires'].quantile(0.945)
dat.loc[(dat['count_of_billionaires'] <= Q55), 'BRN'] = 1
dat.loc[((dat['count_of_billionaires'] > Q55) & (dat['count_of_billionaires'] <= Q79)), 'BRN'] = 2
dat.loc[((dat['count_of_billionaires'] > Q79) & (dat['count_of_billionaires'] <= Q82)), 'BRN'] = 3
dat.loc[((dat['count_of_billionaires'] > Q82) & (dat['count_of_billionaires'] <= Q94)), 'BRN'] = 4
dat.loc[(dat['count_of_billionaires'] > Q94), 'BRN'] = 5

dat.BRN = dat.BRN.astype('int64')
dat.HDSymbol = dat.HDSymbol + "_BRN2"
dat.to_csv('D:\\Michael\\Data\\Ratings\\Billionaire\\BRNEW3.csv', index = False)
dat
```
