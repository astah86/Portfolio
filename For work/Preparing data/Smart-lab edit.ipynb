```python
import pandas as pd
import os
import numpy as np
```

```python
path = 'D:\\Мои документы\\Дивиденд\\Smart-lab\\'
pathnew = 'D:\\Мои документы\\Дивиденд\\Smart-lab_edit\\'

for file in os.listdir(path):

        

        df1 = pd.DataFrame(columns=['index','Дата отчета','Валюта отчета','Выручка, млрд руб','Операционная прибыль, млрд руб','EBITDA, млрд руб',
                                    'Чистая прибыль, млрд руб','Операционный денежный поток, млрд руб','CAPEX, млрд руб','FCF, млрд руб',
                                    'Див.выплата, млрд руб','Дивиденд, руб/акцию','Див доход, ао, %','Дивиденды/прибыль, %',
                                    'Опер. расходы, млрд руб','Себестоимость, млрд руб','Амортизация, млрд руб','НИОКР, млрд руб',
                                    'Расх на персонал, млрд руб','Процентные расходы, млрд руб','Активы, млрд руб','Чистые активы, млрд руб',
                                    'Долг, млрд руб','Наличность, млрд руб','Чистый долг, млрд руб','Цена акции ао, руб','Free Float, %',
                                    'Капитализация, млрд руб','EV, млрд руб','Баланс стоимость, млрд руб','EPS, руб','FCF/акцию, руб',
                                    'BV/акцию, руб','Рентаб EBITDA, %','Чистая рентаб, %','Доходность FCF, %','ROE, %','ROA, %','P/E',
                                    'P/FCF','P/S','P/BV','EV/EBITDA','Долг/EBITDA','R&D/CAPEX, %','CAPEX/Выручка, %'])
        df = pd.read_csv(path + file, sep=";")
        
        df.dropna(axis=0, how='all',inplace=True)
        df.dropna(axis=1, how='all',inplace=True)
        
        try: 
            df.drop(columns=['LTM'], inplace=True)
        except:
            pass
        
        
        
        
        df = df.transpose()
        df.columns = df.iloc[0]
        df = df[1:]
       
        df.reset_index(level=0, inplace=True) 
        for column in df1.columns:
            try:
                df1[column] = df[column]
            except:
                df1[column] = ''
        # df1.fillna(0, inplace=True)
        df1 = df1.replace('%', '', regex=True)
        if file.count('годов') > 0:
            df1['type'] = 'FY'
        if file.count('квартал') > 0:
            df1['type'] = 'Q'
        if file.count('МСФО') > 0:
            df1['report'] = 'MSFO'
        if file.count('РСБУ') > 0:
            df1['report'] = 'RSBU'
        df1['Дата отчета'] = df1['Дата отчета'].astype('datetime64[ns]').dt.strftime('%m/%d/%Y')
        df1.to_excel(pathnew+file[:-4]+'.xlsx',index=False)
        
df1    
```

```python
pathnew1 = 'D:\\Мои документы\\Дивиденд\\Smart-lab_edit1\\'
names = pd.read_excel('D:\\Мои документы\\Дивиденд\\Names.xlsx')
for folder in os.listdir(pathnew1):
    result = pd.DataFrame()
    for file in os.listdir(pathnew1+folder+'\\'):
        df = pd.read_excel(pathnew1+folder+'\\'+file)
        name = file[:file.find('-')-1]
        ticker = names[names['Name'] == name].iloc[0,1]
        df.insert(0, "Ticker", ticker)
        result = pd.concat([result,df], ignore_index=True)
    result.to_excel('D:\\Мои документы\\Дивиденд\\' + folder + '.xlsx', index=False) 
    
result
```
