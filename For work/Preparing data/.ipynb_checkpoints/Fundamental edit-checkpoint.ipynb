```python
pathnew1 = 'D:\\Мои документы\\Дивиденд\\Fundamental_edit1\\'
pathnew = 'D:\\Мои документы\\Дивиденд\\Fundamental_edit\\'
for folder in os.listdir(pathnew):
    for file in os.listdir(pathnew+folder+'\\'):
        df = pd.read_excel(pathnew+folder+'\\'+file)
        df = df.replace('NM',np.NaN)
        df = df.replace('-',np.NaN)
        df = df.replace({'x':''}, regex=True)
        df = df.replace({'%':''}, regex=True)
        
        df = df.replace({'0,0':'0'}, regex=True)
        
        # df['Рын. кап.'] = pd.to_numeric(df['Рын. кап.'])
        for column in df.columns:
            if column != 'Дата оконч. периода':
                df[column] = df[column].astype('string').str.replace('.','')
                df[column] = df[column].astype('string').str.replace(',','.')
                df[column] = df[column].astype('float32')
            else:
                df[column] = df[column].astype('datetime64[ns]').dt.strftime('%m/%d/%Y')
        df.to_excel(pathnew1+folder+'\\'+file, index=False)
        
       
        

df
```
