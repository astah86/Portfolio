```python
import os
import pandas as pd
import numpy as np
import json
import shutil
import datetime as dt
import openpyxl
import sys
from pandas.io.excel import ExcelWriter
from tkinter.messagebox import *
from tkinter.filedialog import *
from PyQt5.QtWidgets import (QFileDialog, QAbstractItemView, QListView,
                             QTreeView, QApplication, QDialog)
class getExistingDirectories(QFileDialog):
    def __init__(self, *args):
        super(getExistingDirectories, self).__init__(*args)
        self.setOption(self.DontUseNativeDialog, True)
        self.setFileMode(self.Directory)
        self.setOption(self.ShowDirsOnly, True)
        self.findChildren(QListView)[0].setSelectionMode(QAbstractItemView.ExtendedSelection)
        self.findChildren(QTreeView)[0].setSelectionMode(QAbstractItemView.ExtendedSelection)
from distutils.dir_util import copy_tree
       
```

```python
qapp = QApplication(sys.argv)
dlg = getExistingDirectories()
if dlg.exec_() == QDialog.Accepted:
    selected_folders = dlg.selectedFiles()
```

```python
Database = input('Enter  group name of portfolios: ')
new_task = 'D://Michael//Data//' + Database
s = '//'
os.makedirs(new_task + s + 'Auto' + s + 'Results')
shutil.copy('D://Portfolio analisis.pbix', new_task + s  + 'Portfolios analysis ' + Database + '.pbix')            
for i in selected_folders:
    folder_path = i.replace("/","//")
    os.mkdir(new_task + s + 'Auto' + s + 'Results' + s + folder_path[folder_path.rfind("/") + 1:])
    copy_tree(folder_path, new_task + s + 'Auto' + s + 'Results' + s + folder_path[folder_path.rfind("/") + 1:])

pathold = new_task + s + 'Auto' + s + 'Results' + s
Sheets = ['Absolute equities', 'ROI equities', 'PercentageEquities', 'InvestedCapital']
Result = pd.DataFrame(columns = ['Date', 'Value', 'Portfolio', 'Sheet', 'Type strategy', 'Risk Rebalancer', 'Re-Entry', 'NotBillionaireMode', 'NA'])
Result1 = pd.DataFrame(columns = ['Date', 'Value', 'Portfolio', 'Sheet', 'Strategy'])
Value = pd.DataFrame(columns = ['Date', 'Value', 'Portfolio', 'Sheet', 'Type strategy', 'Risk Rebalancer', 'Re-Entry', 'NotBillionaireMode', 'NA'])
Value1 = pd.DataFrame(columns = ['Date', 'Value', 'Portfolio', 'Sheet', 'Strategy'])
for pathname in os.listdir(pathold):
    for filename in os.listdir(pathold + pathname + '\\'):
        if (filename.count('Equities')) > 0:
            for i in Sheets:
                Value = Value[0:0]
                Value1 = Value1[0:0]
                Sheet = pd.read_excel(pathold + pathname + '\\' + filename, sheet_name = i)
                Value['Date'] = Sheet['Date']
                Value1['Date'] = Sheet['Date']
                for j in Sheet.columns:
                    
                    if (j!= 'Date'):
                        if ((i == 'PercentageEquities') and (j!='Rebalancer1') and (j[0]!="B")) :
                            Value['Value']  = (Sheet[j] - 1) * 100
                            Value1['Value']  = (Sheet[j] - 1) * 100
                            
                        elif ((i == 'ROI equities') or ((i=='PercentageEquities') and (j=='Rebalancer1')) or((i=='PercentageEquities') and (j[0]=="B"))):
                            Value['Value']  = Sheet[j] * 100
                            Value1['Value']  = Sheet[j] * 100
                        else:
                            Value['Value']  = Sheet[j]
                            Value1['Value']  = Sheet[j]
                        Value['Portfolio'] = pathname
                        Value['Sheet'] = i
                        Value1['Portfolio'] = pathname
                        Value1['Sheet'] = i
                        Value1['Strategy'] = j
                        
                        if (j.count('Origin_') > 0):
                            Value['Type strategy'] = 'Original'
                        elif ((j.count('VqESsiE')) > 0 or (j.count('VQTS') > 0)):
                            Value['Type strategy'] = 'VQ TS'
                        else:
                            Value['Type strategy'] = 'HSI'
                        
                        
                        if (j.count('origin weights') > 0):
                            Value['Risk Rebalancer'] = 'Risk OFF'
                        else:
                            Value['Risk Rebalancer'] = 'Risk ON'
                        
                        
                        if (j.count('_R_') > 0):
                            Value['Re-Entry'] = 'Re-Entry ON'
                        else:
                            Value['Re-Entry'] = 'Re-Entry OFF'
                        
                        
                        if (j.count('_NBM') > 0):
                            Value['NotBillionaireMode'] = 'NBM ON'
                        else:
                            Value['NotBillionaireMode'] = 'NBM OFF'
                        
                        
                        if ((j.count('NA_') > 0) or (j.count('NotAll_') > 0)):
                            Value['NA'] = 'NA ON'
                        else:
                            Value['NA'] = 'NA OFF'
                        
                        
                              
                        Result = pd.concat([Result, Value], ignore_index=True)
                        Result1 = pd.concat([Result1, Value1], ignore_index=True)
                    
                
Result.to_csv(new_task + s + 'Results.csv', index = False)   
Result1.to_csv(new_task + s + 'Results on strategies.csv', index = False)       
Result1
        
```
