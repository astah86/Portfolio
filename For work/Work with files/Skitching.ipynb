# Skitching equity billionaires

```python
path_old = 'C:\\Users\\m.astoshonok\\Documents\\Results\\Results\\'
path_present = 'D:\\Michael\\Data\\Billionaires_May_2023\\Auto\\Results\\'
path_new = 'D:\\Michael\\Data\\Billionaires_May_2023\\Stitched\\'
s = "\\"
for folder in os.listdir(path_old):
    
    os.mkdir(path_new + folder)
    for file in os.listdir(path_old + folder):
        if file.count('equity') > 0 :
            df1 = pd.read_excel(path_old + folder + s + file, sheet_name = 'PercentageEquities', parse_dates = [0])
            df1 = df1.iloc[:, 0:3]
            break
        if file.count('chart') > 0 :
            df1 = pd.read_excel(path_old + folder + s + file, sheet_name = 'PercentageEquities', parse_dates = [0])
            df1 = df1.iloc[:, 0:3]
            df1.iloc[:, 1:3] = df1.iloc[:, 1:3] + 1
            break

    for files in os.listdir(path_present + folder):
        if files.count('Equities') > 0 or files.count('curves') > 0:
            df = pd.read_excel(path_present + folder + s + files, sheet_name = 'PercentageEquities', parse_dates = [0])

            df = df[df1.columns]
            break

    for i in [1,2]:
        a = df.iloc[0:,i].shift(-1) / df.iloc[0:,i] 
        a = a[:len(a)-1]
        df.iloc[1:,i] = a
    last_date = np.unique(df1['Date'])[len(np.unique(df1['Date']))-1]
    df = df[df['Date'] >= last_date]
    df2 = df1.merge(df, how='outer', on='Date')
    for j in range(len(df1), len(df2)):
        for k in [1,2]:
            df2.iloc[j, k] = df2.iloc[j-1, k] * df2.iloc[j, k+2]
    df2 = df2.iloc[:, :3]
    df2.Date = df2.Date.astype('datetime64').dt.strftime("%m/%d/%Y")
    df2.to_excel(path_new + folder + s + folder + ' equity curves' + '.xlsx', index=False)
    
        
    
```

```python
path_old = 'C:\\Users\\m.astoshonok\\Documents\\Results Billionaires_Jan_2022\\Results Billionaires_Jan_2022\\'
path_old_b = 'C:\\Users\\m.astoshonok\\Documents\\Results\\Results\\'
path_present = 'D:\\Michael\\Data\\Billionaires_May_2023\\Auto\\Results\\'
path_new = 'D:\\Michael\\Data\\Billionaires_May_2023\\Stitched\\'
s = "\\"
for folder in os.listdir(path_old):
    
    for file in os.listdir(path_old + folder):
        if file.count('rade') > 0:
            origin_old = pd.read_excel(path_old + folder + s + file, sheet_name = 'Origin_origin weights', parse_dates = [0])
            last_date = np.unique(origin_old['Exit Date'])[len(np.unique(origin_old['Exit Date']))-1]
            origin_old = origin_old[origin_old['Exit Date'] < last_date]
            break
    
    for files in os.listdir(path_present + folder):
        if files.count('rade') > 0:
            origin_present = pd.read_excel(path_present + folder + s + files, sheet_name = 'Origin_origin weights', parse_dates = [0])
            origin_present = origin_present[origin_present['Exit Date'] >= last_date]
            break
            

    for file in os.listdir(path_old_b + folder):
        if file.count('equity') > 0 or file.count('chart') > 0:
            df1 = pd.read_excel(path_old_b + folder + s + file, sheet_name = 'PercentageEquities', parse_dates = [0])
            df1 = df1.iloc[:, 0:3]
            strategy = df1.columns[2]
            
    for file in os.listdir(path_old + folder):
        if file.count('rade') > 0:
            billi_old = pd.read_excel(path_old + folder + s + file, sheet_name = strategy, parse_dates = [0])
            last_date_b = np.unique(billi_old['Exit Date'])[len(np.unique(billi_old['Exit Date']))-1]
            billi_old = billi_old[billi_old['Exit Date'] < last_date_b]
            
            break
    for files in os.listdir(path_present + folder):
        if files.count('rade') > 0:
            billi_present = pd.read_excel(path_present + folder + s + files, sheet_name = strategy, parse_dates = [0])
            billi_present = billi_present[billi_present['Exit Date'] >= last_date_b]
            break
    
    possize_present = billi_present[billi_present['Exit Date'] == last_date_b]['Position Value'].sum()
    last_date_b = np.unique(billi_old['Exit Date'])[len(np.unique(billi_old['Exit Date']))-1]
    possize_old = (billi_old[billi_old['Exit Date'] == last_date_b]['Position Value'] * \
                                    (1 + billi_old[billi_old['Exit Date'] == last_date_b]['% Profit'] / 100)).sum()
    billi_present['Position Value'] = possize_old * billi_present['Position Value']/possize_present 
    billi_present['Shares'] = billi_present['Position Value'] / billi_present['Exit Price']
    billi_present['Absolute Profit'] = billi_present['Position Value'] * (billi_present['% Profit'] / 100)
    
    
    origin_new = origin_old.append(origin_present)
    billi_new = billi_old.append(billi_present)
    origin_new['Entry Date'] = origin_new['Entry Date'].astype('datetime64').dt.strftime("%m/%d/%Y")
    origin_new['Exit Date'] = origin_new['Exit Date'].astype('datetime64').dt.strftime("%m/%d/%Y")
    billi_new['Entry Date'] = billi_new['Entry Date'].astype('datetime64').dt.strftime("%m/%d/%Y")
    billi_new['Exit Date'] = billi_new['Exit Date'].astype('datetime64').dt.strftime("%m/%d/%Y")
    writer = pd.ExcelWriter(path_new + folder + s + folder + ' trade details' + '.xlsx', engine = 'xlsxwriter')
    origin_new.to_excel(writer, sheet_name = 'Origin_origin weights', index=False)
    billi_new.to_excel(writer, sheet_name = strategy, index=False)
    writer.close()
    
          
                   
    
#     for files in os.listdir(path_present + folder):
#         if files.count('Equities') > 0 or files.count('curves') > 0:
#             df = pd.read_excel(path_present + folder + s + files, sheet_name = 'PercentageEquities', parse_dates = [0])
            
#             df = df[df1.columns]
    
#     for i in [1,2]:
#         a = df.iloc[0:,i].shift(-1) / df.iloc[0:,i] 
#         a = a[:len(a)-1]
#         df.iloc[1:,i] = a
#     df2 = df1.merge(df, how='outer', on='Date')
#     for j in range(len(df1), len(df2)):
#         for k in [1,2]:
#             df2.iloc[j, k] = df2.iloc[j-1, k] * df2.iloc[j, k+2]
#     df2 = df2.iloc[:, :3]
#     df2.Date = df2.Date.astype('datetime64').dt.strftime("%m/%d/%Y")
#     df2.to_excel(path_new + folder + '.xlsx', index=False)
billi_new
```
