```python
import os
import pandas as pd
import numpy as np
import json
import shutil
import datetime as dt
import openpyxl
from openpyxl import load_workbook
import sys
import docx
from tkinter import Tk, filedialog
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import Pt, RGBColor
from pandas.io.excel import ExcelWriter
```

```python
billi_2022 = 'D:\\Michael\\Data\\Billionaires-May-2024\\Results Billionaires_Jan_2022\\'
billi_2024 = 'D:\\Michael\\Data\\Billionaires-May-2024\\Results Billionaires_May_2024\\'
stitched = 'D:\\Michael\\Data\\Billionaires-May-2024\\StitchedAuto\\'
# col = ['Date','Origin_origin weights','Origin_VQ pos size','Billionaires2','Billionaires5','Billionaires5M']
col = ['Date','Origin_origin weights','Billionaires2','Billionaires5','Billionaires5M']
# col = ['Date','Origin_origin weights','Billionaires5M']
for billi in os.listdir(billi_2022):
    print(billi)
    Equities = pd.DataFrame()
    
    
    for strategy in col[1:]:
        print(strategy)
        Equity = pd.DataFrame()
        TradeDetails = pd.DataFrame()
        value_old = 0
        
        for file in os.listdir(billi_2022 + billi + '\\'):
       
            if file.count('Equities') > 0:
                Equity_2022 = pd.read_excel(billi_2022 + billi + '\\' + file, sheet_name = 'PercentageEquities', usecols = ['Date', strategy], parse_dates = [0])
    
            if file.count('TradeDetails') > 0:
                TradeDetails_2022 = pd.read_excel(billi_2022 + billi + '\\' + file, sheet_name = strategy)
                
    
        for file in os.listdir(billi_2024 + billi + '\\'):
            if file.count('Equities') > 0:
                Equity_2024 = pd.read_excel(billi_2024 + billi + '\\' + file, sheet_name = 'PercentageEquities', usecols = ['Date', strategy], parse_dates = [0])
                if strategy.count('Billi') > 0:
                    Equity_2024[strategy] = Equity_2024[strategy] + 1
            if file.count('TradeDetails') > 0:
                TradeDetails_2024 = pd.read_excel(billi_2024 + billi + '\\' + file, sheet_name = strategy)
                

        TradeDetails_2022['Entry Date'] = TradeDetails_2022['Entry Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
        TradeDetails_2022['Exit Date'] = TradeDetails_2022['Exit Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
        TradeDetails_2024['Entry Date'] = TradeDetails_2024['Entry Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
        TradeDetails_2024['Exit Date'] = TradeDetails_2024['Exit Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
        
        dates_2022_O  = pd.unique(TradeDetails_2022['Entry Date'])
        dates_2022_O = sorted((dt.datetime.strptime(x, '%m/%d/%Y') for x in dates_2022_O), reverse=True)
        dates_2024_O  = pd.unique(TradeDetails_2024['Entry Date'])
        
        for date in dates_2022_O[1:]:
            close_date = pd.unique(TradeDetails_2022[TradeDetails_2022['Entry Date'] == date.strftime('%m/%d/%Y')]['Exit Date'])
            
            if (len(close_date) == 1 and close_date[0] in dates_2024_O) and close_date[0] != TradeDetails_2022.iloc[0, 4]:
                print(close_date[0])
                break
            if close_date[0] == TradeDetails_2022.iloc[0, 4]:
                close_date = pd.unique(TradeDetails_2022[TradeDetails_2022['Entry Date'] == dates_2022_O[1].strftime('%m/%d/%Y')]['Exit Date'])
                close_date[0] = close_date[len(close_date)-1]
                print(close_date[0])
                
        Equity_2022  = Equity_2022[Equity_2022['Date'].astype('datetime64[ns]') <= dt.datetime.strptime(close_date[0], '%m/%d/%Y')]
        Equity_2024  = Equity_2024[Equity_2024['Date'].astype('datetime64[ns]') >= dt.datetime.strptime(close_date[0], '%m/%d/%Y')]
        TradeDetails_2022 = TradeDetails_2022[TradeDetails_2022['Exit Date'].astype('datetime64[ns]') <= dt.datetime.strptime(close_date[0], '%m/%d/%Y')]  
        TradeDetails_2024 = TradeDetails_2024[TradeDetails_2024['Entry Date'].astype('datetime64[ns]') >= dt.datetime.strptime(close_date[0], '%m/%d/%Y')] 
        value_old = Equity_2022[Equity_2022['Date'].astype('datetime64[ns]') == dt.datetime.strptime(close_date[0], '%m/%d/%Y')][strategy]

        Equity_2024['calculation'] = Equity_2024[strategy] / Equity_2024[strategy].shift(1)
        Equity_2024.iloc[0,2] = value_old
        Equity_2024['calculation2'] = Equity_2024['calculation'].cumprod()
        Equity_2024 = Equity_2024.iloc [1: , :]
        Equity_2024[strategy] = Equity_2024['calculation2']
        Equity_2024.drop(columns=['calculation', 'calculation2'], inplace=True)
        if strategy == 'Origin_origin weights':
            TradeDetails = pd.concat([TradeDetails_2022, TradeDetails_2024], ignore_index=True)
            os.mkdir(stitched + billi)
            TradeDetails.to_excel(stitched + billi + '\\TradeDetails.xlsx', index=False, sheet_name=strategy)
            Equities = pd.concat([Equity_2022, Equity_2024], ignore_index=True)
            
        
        else:
            position_old = TradeDetails_2022[TradeDetails_2022['Exit Date'].astype('datetime64[ns]') == dt.datetime.strptime(close_date[0], '%m/%d/%Y')]['Position Value'].sum() 
            position_new = TradeDetails_2024[TradeDetails_2024['Entry Date'].astype('datetime64[ns]') == dt.datetime.strptime(close_date[0], '%m/%d/%Y')]['Position Value'].sum()
            k = position_old / position_new 
            TradeDetails_2024['Position Value'] = TradeDetails_2024['Position Value'] * k
            TradeDetails_2024['Shares'] = TradeDetails_2024['Position Value'] / TradeDetails_2024['Entry Price'] 
            TradeDetails_2024['Absolute Profit'] = TradeDetails_2024['Position Value'] * TradeDetails_2024['% Profit'] / 100
            TradeDetails = pd.concat([TradeDetails_2022, TradeDetails_2024], ignore_index=True)
            Equity = pd.concat([Equity_2022, Equity_2024], ignore_index=True)
            Equities[strategy] = Equity[strategy]

            
            with pd.ExcelWriter(stitched + billi + '\\TradeDetails.xlsx', engine='openpyxl', mode='a') as writer:  
                TradeDetails.to_excel(writer, sheet_name=strategy, index=False)
    Equities['Date']  = Equities['Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
    Equities.to_excel(stitched + billi + '\\Equities.xlsx', index=False, sheet_name='PercentageEquities') 
    
    
    
            
           
        
        
       
   
```
