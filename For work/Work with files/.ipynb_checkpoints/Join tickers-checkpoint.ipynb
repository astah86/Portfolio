# Making tickers

```python
def out_red(text):
    print("\033[91m")
    print( "\033[1m {}" .format(text))
    print( "\033[0m" )


def findCap(s):
    string = ''
    for ele in str(s):
        if ord(ele) >= 65 and ord(ele) <= 90:
            string += ele
    return string
mainpath = 'D:\\Michael\\Data\\'
Name = Database = input('Enter  group name of portfolios: ')
ID = mainpath + Name + '\\Input Data\\'
pathtickers = mainpath + Name + '\\Tickers\\'
os.mkdir(pathtickers)
for file in os.listdir(ID):
    tickers = pd.DataFrame()
    IDfile = pd.read_csv(ID + file, parse_dates = ['Open Date', 'Close Date'], dayfirst = False, sep = ",")
    IDfile = IDfile[IDfile['Position Type'] != 'Combined']
    if (IDfile.shape[1] < 500):
        IDfile.drop(IDfile[(IDfile.Symbol.map(len) > 15) | (IDfile['Exchange Market'] == 'Options Pric') | (IDfile['Trade Group'].str.lower() == 'options')].index, inplace=True)
        tickers['Symbol']= IDfile.Symbol
        tickers['Type']= IDfile.Type
        tickers = tickers.replace({'Type':{'Buy Stock':'Long'}})
#         tickers = tickers.replace({'Symbol':{'/':'-'}})#only for crypto
        tickers['Weight']= np.abs(IDfile.Weight)
        tickers['Open Date']= IDfile['Open Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
       
        tickers['Open Price']= IDfile['Open Price']
        tickers['Close Date']= IDfile['Close Date'].astype('datetime64[ns]').dt.strftime("%m/%d/%Y")
        tickers['Close Price']= IDfile['Close Price']
        tickers['Position size'] = 1000 * tickers['Weight']
        tickers['Weight'] = 1
        tickers['Multiplier'] = 1
        
        
        tickers.dropna(subset=['Type'], inplace=True)

        namefile = file[:file.find("_")]
        newname = findCap(namefile)
        
        tickers.to_csv(pathtickers + newname + '.csv', index = False)
        
    else: 
        out_red("Something went wrong with file " + file + ", you have to look with your eyes" )
    

```

```python
# Bcount = Bcount.drop(columns = ['nameOfIssuer'])
Bcount = Bcount.sort_values(['tickercusip', 'date'])
Bcount
```

```python
Bcount['NetBilli'] = [2*Bcount.iloc[n, 2] - Bcount.iloc[n-1, 2] if ((n != 0) and (Bcount.iloc[n, 2] < Bcount.iloc[n-1, 2]) and (Bcount.iloc[n, 0] == Bcount.iloc[n-1, 0]))
                  else  Bcount.iloc[n, 2] for n in range(0, Bcount.shape[0])]
Bcount
```

```python
Bcount['Ex'] = Bcount.cik
Bcount.loc[((Bcount['cik'] < Bcount['cik'].shift()) & (Bcount['tickercusip'] == Bcount['tickercusip'].shift())), 'Ex'] = 2*Bcount.cik - Bcount.cik.shift()
```

```python
Bcount['Check'] = Bcount.NetBilli - Bcount.Ex
Bcount
```

```python
Bcount = Bcount.drop(columns = ['cik', 'NetBilli', 'Check'])
Bcount
```

```python
Neg = Bcount[Bcount.Ex < 0]
Pos = Bcount[Bcount.Ex > 0]
Neg = Neg.drop_duplicates(subset=['date', 'Ex'])
Pos = Pos.drop_duplicates(subset=['date', 'Ex'])
Neg
```

```python

Pos = Pos.groupby(['date']).median()
Neg = Neg.groupby(['date']).median()
Neg
```

```python
Pos.insert(0,'date',Pos.index)
Neg.insert(0,'date',Neg.index)
Neg
```

```python
Pos.rename(columns={'Ex': 'PosMed'}, inplace=True)
Neg.rename(columns={'Ex': 'NegMed'}, inplace=True)
Pos
```

```python
Pos = Pos.reset_index(drop=True)
Neg = Neg.reset_index(drop=True)
Pos
```

```python


BR = pd.merge(Bcount, Pos)
BR = pd.merge(BR, Neg)
BR
```

```python
BR.loc[(BR['Ex'] < BR['NegMed']), 'BRFMP'] = 1
BR.loc[((BR['Ex'] >= BR['NegMed']) & (BR['Ex'] < 0)), 'BRFMP'] = 2
BR.loc[(BR['Ex'] == 0), 'BRFMP'] = 3
BR.loc[((BR['Ex'] <= BR['PosMed']) & (BR['Ex'] > 0)), 'BRFMP'] = 4
BR.loc[(BR['Ex'] > BR['PosMed']), 'BRFMP'] = 5

BR = BR.drop(columns = ['PosMed', 'NegMed'])
BR.BRFMP = BR.BRFMP.astype('int64')
BR.tickercusip = BR.tickercusip + "_BRFMP"
BR.to_csv('D:\\Michael\\Data\\Ratings\\Billionaire\\BRFMP.csv', index = False)
BR
```

```python
def my_basic_research(df, column):
    print("Базовые метрики")
    print(df[column].describe())
    print("------------------------------------")
    
    print("Самые популярные значения метрики, топ 5")
    print(df[column].value_counts().nlargest(5))
    print("------------------------------------")
    
    print("Эксцесс ", kurtosis(df[column]))
    print("Ассиметрия ", skew(df[column]))
    
    sns.distplot(df[column])
    plt.title("Распределение метрики")
my_basic_research(BR, 'BRFMP')
```
