```python
pathold = 'D:\\Desktop\\New folder (2)\\26B\\'
pathnew = 'D:\\Desktop\\New folder (2)\\27Bi\\'
Result = pd.DataFrame()
n = 0
k = 0
for filename in os.listdir(pathold):
    
    dat = pd.read_csv(pathold + filename, parse_dates = [' Date open'], dayfirst = False)
    a = dat.iloc[:,7].max()
    count0 = dat.iloc[:,0].count()
    for filename1 in os.listdir(pathnew):
        if (filename1.count(filename)) > 0:
            
            dat1 = pd.read_csv(pathnew + filename1, parse_dates = ['FilingOpenDate'], dayfirst = False)
            b = dat1.iloc[:,7].max()
            count1 = dat1.iloc[:,0].count()
            fd2 = dat1.iloc[1,3]
            fn = filename1
            
    
    if (a/b>2):
        Result.loc[n,'Name'] = filename[:-4]
        Result.loc[n,'Name2'] = fn[:-4]
        Result.loc[n,'PositionSize'] = 'Extrimal small'
        
        
        
    elif (a/b<0.5):
        Result.loc[n,'Name'] = filename[:-4]
        Result.loc[n,'Name2'] = fn[:-4]
        
        Result.loc[n,'PositionSize'] = 'Extrimal large'
        
        
    else:
        Result.loc[n,'Name'] = filename[:-4]
        Result.loc[n,'Name2'] = fn[:-4]
        Result.loc[n,'PositionSize'] = 'normal'
        
    Result.loc[n, 'Ratio_PosSize'] = a/b
    Result.loc[n, 'absolut max'] = b
    
    Result.loc[n, 'old_count'] = count0
    Result.loc[n, 'new_count'] = count1
    Result.loc[n, 'Diff count'] = count0/count1
    Result.loc[n, 'first_date1'] = dat.iloc[1,3]
    Result.loc[n, 'first_date2'] = fd2
    n += 1
    
    
Result['Diff_days'] = Result['first_date1']-Result['first_date2']
#Result.sort_values(['PositionSize', 'Diff'], ascending = [True, False])
Result.sort_values('Diff_days', ascending = True)
#Result.to_csv('C:\\Users\\Msi-1\\Desktop\\New folder (2)\\For analyse.csv', index_label = 'index')
```

```python
pathold = 'D:\\Michael\\Data\\Billionaires_Jan_2022\\Old tickers\\'
pathnew = 'D:\\Michael\\Data\\Billionaires_Jan_2022\\1_17_2022 Billionaires Data\\Tickers\\'
Match = pd.DataFrame()
n = 0
for filename in os.listdir(pathold):
    
        
    dat = pd.read_csv(pathold + filename, parse_dates = [' Date open'], dayfirst = False)
    k = 0
    k1 = 0
    dat = dat[dat[' Date open'] > '1/1/2014']


    for filename1 in os.listdir(pathnew):
        if (filename1.count(filename[:-4])) > 0:
            dat1 = pd.read_csv(pathnew + filename1, parse_dates = ['FilingOpenDate'], dayfirst = False)
            dat1['FilingOpenDate'] = pd.to_datetime(dat1['FilingOpenDate']).dt.normalize()
            dat1 = dat1[dat1['FilingOpenDate'] < '1/1/2021']
            dat1 = dat1 [dat1['FilingOpenDate'] > '1/1/2014']
            fn = filename1
    dat['YearMonth'] = dat[' Date open'].map(lambda x: 100*x.year + x.month)
    dat1['YearMonth'] = dat1['FilingOpenDate'].map(lambda x: 100*x.year + x.month)
    dat['General'] = dat.iloc[:,0].astype('str') + dat.iloc[:,10].astype('str') 
    dat1['General'] = dat1.iloc[:,0].astype('str') + dat1.iloc[:,9].astype('str') 
    dat['General2'] = dat.iloc[:,0].astype('str') + dat.iloc[:,10].astype('str') + dat.iloc[:,7].astype('int').astype('str')
    dat1['General2'] = dat1.iloc[:,0].astype('str') + dat1.iloc[:,9].astype('str') + dat1.iloc[:,7].astype('int').astype('str')
    
    for i in range(dat['General'].count()):

        if list(dat1['General']).count(dat.iloc[i, 11]) > 0:
            k += 1
    for i in range(dat['General2'].count()):

        if list(dat1['General2']).count(dat.iloc[i, 12]) > 0:
            k1 += 1
    Match.loc[n,'Name'] = filename[:-4]
    Match.loc[n,'Name2'] = fn[:-4]
    Match.loc[n,'Count_ old trades'] = dat.shape[0]
    Match.loc[n,'Count_ new trades'] = dat1.shape[0]
    Match.loc[n,'Count_match'] = k
    Match.loc[n,'Count_match_PS'] = k1
        
   
    
    n += 1
    print('Billi = ', n)
Match['Percent'] = Match['Count_match'] * 100 / Match['Count_ old trades']
Match['Percent_PS'] = Match['Count_match_PS'] * 100 / Match['Count_ old trades']
Match.to_csv('D:\\Michael\\Data\\13f\\MatchHD.csv', index_label = 'index')
Match.sort_values('Percent_PS')
```

```python
pathresult = 'D:\\Michael\\Data\\PortPhilip_03_22(Australia)\\Auto\\Results\\'
strategy = pd.read_csv('D:\\Michael\\Data\\PortPhilip_03_22(Australia)\\Auto\\PortfolioStrategy.csv', index_col = 'Portfolio')
new = 'D:\\Michael\\Data\\PortPhilip_03_22(Australia)\\Auto\\For Legals\\'
trades = 'Individual Trades'
os.mkdir(new)
for pathname in os.listdir(pathresult):
    os.mkdir(new + pathname)
    os.mkdir(new + pathname + '\\' + trades)
    Equity = pd.DataFrame()
    for filename in os.listdir(pathresult + '\\' + pathname):
        if ((filename[-4:] == 'rwch') and (filename.count('_') > 0)) or ((filename[-3:] == 'png') and (filename.count('_') > 0)) or ((filename[-4:] == 'xlsx') and (filename.count('stop') > 0)):
            shutil.copy(pathresult + '\\' + pathname + '\\' + filename, new + pathname + '\\' + trades + '\\' + filename)
        if ((filename[-3:] == 'png') and (filename.count('_') == 0)) :
            shutil.copy(pathresult + '\\' + pathname + '\\' + filename, new + pathname  + '\\' + pathname + '.png')
        if ((filename[-4:] == 'rwch') and (filename.count('_') == 0)) :
            shutil.copy(pathresult + '\\' + pathname + '\\' + filename, new + pathname  + '\\' + pathname + '.rwch')
        if ((filename[-4:] == 'xlsx') and (filename.count('Equities') > 0)) :
            Equity = pd.read_excel(pathresult + pathname + '\\' + filename, sheet_name = strategy.loc[pathname,'Sheet'], usecols = ['Date', 'Origin_origin weights'])
            tradesmith = pd.read_excel(pathresult + pathname + '\\' + filename, sheet_name = 'PercentageEquities', usecols = [strategy.loc[pathname,'Strategy']])
            Equity.insert(2, strategy.loc[pathname,'Strategy'], tradesmith)
            Equity['Date'] = Equity['Date'].dt.strftime("%m/%d/%Y")
            Equity.to_excel(new + pathname  + '\\' + pathname + ' equity curves.xlsx', index = False)
        if ((filename[-4:] == 'xlsx') and (filename.count('TradeDetails') > 0)):
            tradedetails = new + pathname + '\\' + pathname + ' trade details.xlsx'
            shutil.copy(pathresult + '\\' + pathname + '\\' + filename, tradedetails)
            wb = openpyxl.load_workbook(tradedetails)
            sheet = wb.sheetnames
            for i in sheet:
                if (i != 'Origin_origin weights') and (i != strategy.loc[pathname,'Strategy']):
                    pfd = wb[i]
                    wb.remove(pfd)
            wb.save(tradedetails)
    
```
