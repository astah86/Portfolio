```python
def inch_to_cm(value):
    return value / 2.54

def findCap(s):
    string = ''
    for ele in str(s):
        if ord(ele) >= 65 and ord(ele) <= 90:
            string += ele
    return string
s = '//'

today = dt.date.today().strftime('%m/%d/%Y')
openkey = 0
for pathname in os.listdir(folder_png):
    list_files = os.listdir(folder_png + s + pathname)
    n = 0
    line = strat[strat.Portfolio == pathname].iloc[0,1]
    lineoriginal = line.replace('VQ pos size','origin weights')
    
    tickers = pd.read_csv(folder_png[: folder_png.find('Auto')] + 'Tickers//' + findCap(pathname) + '.csv')
    Count_tickers = tickers.iloc[:,0].count()
    
    for filename in os.listdir(folder_png + s + pathname):
        if filename.count('TradeDetails') > 0:
            trades = pd.read_excel(folder_png + s + pathname + s + filename, sheet_name = 'Origin_origin weights')
            Count_trades = trades.iloc[:,0].count()
    
    trades_processed = round(Count_trades * 100 / Count_tickers,1)

    while n < len(list_files):
        
        if ((list_files[n].count('_') == 0) and list_files[n].count('png') > 0) :
            
            if openkey == 0:
                
                doc = docx.Document()
                doc.save(folder_png[:folder_png.rfind("/") + 1] + folder_png[folder_png.find('Data//') + 6: folder_png.find('//Auto') ]  + '_Charts' +  '.docx')
               
                doc = docx.Document(folder_png[:folder_png.rfind("/") + 1] + folder_png[folder_png.find('Data//') + 6: folder_png.find('//Auto') ]  + '_Charts' +  '.docx')
                
                title = doc.add_paragraph(' backtest with Tradestops strategies')
                title.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                title.style = doc.styles['Heading 1']
                
                author = doc.add_paragraph('Author: Mikhail Astoshonok')
                author.alignment = WD_ALIGN_PARAGRAPH.LEFT
                author.style = doc.styles['Normal']
                
                date = doc.add_paragraph(f'Date of creation: {today}')
                date.alignment = WD_ALIGN_PARAGRAPH.LEFT
                date.style = doc.styles['Normal']
                
                request = doc.add_paragraph('Request:')
                request.alignment = WD_ALIGN_PARAGRAPH.LEFT
                request.style = doc.styles['Heading 4']
                
                backtest_results = doc.add_paragraph('Back test results:')
                backtest_results.alignment = WD_ALIGN_PARAGRAPH.LEFT
                backtest_results.style = doc.styles['Heading 3']
                
                p = doc.add_paragraph('Portfolio name: ' + pathname + ' (' + str(trades_processed) + '% trades_processed)')
                p.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                p.style = doc.styles['Heading 1']
                               
                doc.add_picture(folder_png + s + pathname + s + list_files[n], width=docx.shared.Inches(inch_to_cm(13.330)), height=docx.shared.Inches(inch_to_cm(9)))
                p1 = doc.add_paragraph(f'The strategy used is {line}')
                p1.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                p1.style = doc.styles['Normal']
                if (str(list_files).count('$') > 0):
                    doc.add_picture(folder_png + s + pathname + s + str(list_files[n])[:-4] + '_$.png', width=docx.shared.Inches(inch_to_cm(13.330)), height=docx.shared.Inches(inch_to_cm(9)))
                    p1 = doc.add_paragraph(lineoriginal)
                    p1.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                    p1.style = doc.styles['Normal']
                openkey = 1
            else:
                p = doc.add_paragraph('Portfolio name: ' + pathname + ' (' + str(trades_processed) + '% trades_processed)')
                p.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                p.style = doc.styles['Heading 1']
                doc.add_picture(folder_png + s + pathname  + s + list_files[n], width=docx.shared.Inches(inch_to_cm(13.330)), height=docx.shared.Inches(inch_to_cm(9)))
                p1 = doc.add_paragraph(f'The strategy used is {line}')
                p1.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                p1.style = doc.styles['Normal']
                if (str(list_files).count('$') > 0):
                    print(str(list_files))
                    doc.add_picture(folder_png + s + pathname + s + str(list_files[n])[:-4] + '_$.png', width=docx.shared.Inches(inch_to_cm(13.330)), height=docx.shared.Inches(inch_to_cm(9)))
                    p1 = doc.add_paragraph(lineoriginal)
                    p1.alignment = WD_ALIGN_PARAGRAPH.CENTER 
                    p1.style = doc.styles['Normal']
            break
        n += 1
    p2 = doc.add_paragraph('Trades examples that helped to overperform the portfolio: ')
    p2.style = doc.styles['Heading 3']
    p2.alignment = WD_ALIGN_PARAGRAPH.LEFT 
   
    trades = 0
    for png in list_files:
        if (png.count('_') > 0) and (png.count('png') > 0) and (png.count('$') == 0):
            doc.add_picture(folder_png + s + pathname + s + png, width=docx.shared.Inches(inch_to_cm(13.330)), height=docx.shared.Inches(inch_to_cm(9)))
            trades = 1
    if trades == 0:
        p2 = doc.add_paragraph('No trades examples that helped to overperform the portfolio: ')
        p2.style = doc.styles['Heading 3']
        p2.alignment = WD_ALIGN_PARAGRAPH.LEFT
p.style.font.size = Pt(14)
p2.style.font.size = Pt(12) 
p1.style.font.size = Pt(10)       
doc.save(folder_png[:folder_png.rfind("/") + 1] + folder_png[folder_png.find('Data//') + 6: folder_png.find('//Auto') ]  + '_Charts' +  '.docx')            
    
```
