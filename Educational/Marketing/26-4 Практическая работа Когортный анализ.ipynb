```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt 
```

```python
data = pd.read_excel('26.4 Когортный анализ.xlsx')
data.head()
```

```python
data.info()
```

```python
def get_month(x):
    return dt.datetime(x.year,x.month,1)
```

```python
data['month_year']=data['date'].apply(get_month)
data['cohortMonth']=data.groupby('user_id')['month_year'].transform('min')
data.head(50)
```

```python
def get_date(df,column):
    year=df[column].dt.year
    month=df[column].dt.month
    day=df[column].dt.day
    return year,month,day
```

```python
date_year, date_month, _ = get_date(data,'date')
cohort_year ,cohort_month , _=get_date(data,'cohortMonth')
year_diff = date_year - cohort_year
month_diff = date_month - cohort_month
data['cohort_index'] = year_diff * 12 + month_diff + 1
data.head()
```

```python
data1=data.drop_duplicates(subset = ['user_id', 'cohortMonth', 'cohort_index']).pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='user_id',
                      aggfunc='count')
data1
```

```python
first_column = data1.iloc[:,0]
retention = data1.div(first_column,axis=0)
retention.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ-retention')
sns.heatmap(data=retention,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

```python
data2=data.pivot_table(index='cohortMonth',
                      columns='cohort_index',
                      values='price',
                      aggfunc=sum)
data2
```

```python
first_column1 = data2.iloc[:,0]
retention1 = data2.div(first_column1,axis=0)
retention1.round(2)
```

```python
plt.figure(figsize=(11,9))
plt.title('когортный анализ-retention')
sns.heatmap(data=retention1,
           annot=True,
           fmt='.0%',
           vmin=0.0,
           vmax=0.5,
           cmap="YlGnBu")
plt.show()
```

Из проведенного выше анализа можно заметить, что самое большое падение из новых клиентов происходит в первый месяц, при этом сумма покупок оставшихся клиентов через один месяц может даже превышать сумму покупок в первый месяц. Также можно заметить, что возвращаемость клинтов пришедших в четные месацы выше, тем тех кто пришли в нечетные. Это может говорить о некоторой сезонности или переодической отсутвии/наличии каких нибудь акций и т.п. В последние месяцы произошла заметное снижение покупательской активности особенно среди тех, кто пришел в феврале 20-го года. Это может говорить о снижении спроса на товары, которые были актуальны долгое время с февраля 20-го года. Это следовало бы изучить более детально.
