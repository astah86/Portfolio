# Задание:

**Цель работы**

Доказать вероятности для разных стратегий, которые получаются в парадоксе Монти Холла.

**Что нужно сделать**

В рамках данной практической работы необходимо:
- Доказать вероятность для стратегии стратегия **STAY** ($\frac{1}{3}$)
- Доказать вероятность для стратегии стратегия **RANDOM CHOICE** ($\frac{1}{2}$)
- Доказать вероятность для стратегии стратегия **SWITCH** ($\frac{2}{3}$)

Результатом практической работы является доказательство с пояснениями.

# Решение: 

**Напомним задачу**:

Перед участником три закрытые двери.
- За одной дверью новый автомобиль.
- За двумя другими - козы.

Цель участника - выиграть автомобиль!

Участник сначала выбирал одну дверь. Но ведущий не говорил, что за ней находится, а открывал другую дверь, за которой находилась коза. И задавал вопрос - "не хотите ли вы изсенить выбор"?

Участник должен был решить - стоит ли менять свой выбор или это психологический ход?

С какой вероятностью вы выиграете автомобиль, если:
- Не будете менять свой выбор (стратегия STAY)
- Изменить выбор случайным образом (например, подкинуть монетку) (стратегия RANDOM CHOICE)
- Изменить свой выбор на противоположный (стратегия SWITCH)

**Предположения, в которых мы решаем задачу**:
- автомобиль размещен за любой из дверей с одинаковой вероятнотсть;
- ведущий знает, где находится автомобиль;
- ведущий в любом случае открывает дверь с козой (но не ту, которую выбрал игрок) и предлагает игроку изменить выбор;
- если у ведущего есть выбор, то он выбирает любую из двух дверей с одинаковой вероятностью.

**Первая стратегия (STAY):**     

Если вы не меняете свой выбор и всегда открываете ту дверь, которую выбрали в самом начале, то победите только в то случае, если в самом начале выбрали дверь с автомобилем. Такая дверь одна, а всего дверей три, то вероятность выигрыша составит 1/3.

**Вторая стратегия (RANDOM CHOICE):**      

После того, как ведущий открыл дверь, у нас остается 2 двери. За одной коза, а за другой автомобиль. Вероятность выигрыша в этом случае составит 1/2.

**Третья стратегия (SWITCH):**     

В третьей стратегии мы всегда меняем свой выбор на противоположный. Поэтому проиграть мы можем только в том случае, если с самого начала выиграли автомобиль. Если мы действительно в самом начале выбрали авто, то после того, как ведущий открыл дверь с козой, мы меняем свой выбор на противоположный и выбираем козу.

С другой стороны, мы выиграем во всех остальных случаях. Если мы в самом начале выбираем козу, а ведущий открывает другую дверь с козой, то мы меняем свой выбор и выигрываем автомобиль. Дверей с козами две, а всего дверей три. Вероятность выигрыша в этом случае составит 2/3.

В результате мы получаем следующие вероятности для каждой из стратегий:
- **стратегия STAY - 1/3**
- **стратегия RANDOM CHOICE - 1/2**
- **стратегия SWITCH - 2/3**
