```python
from sympy import *
from sympy.plotting import plot3d
import math
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
f = 0.8 * x + 20 * y
f
```

```python
plot3d(f)
```

```python
f = x * y / sqrt(x**2 + y**2 - 5)
f
```

```python
plot3d(f)
```

```python
f = sin(sqrt(x ** 2 + y ** 2))
f
```

```python
plot3d(f)
```

```python
f = x*cos(y)
f
```

```python
plot3d(f)
```

```python
f = math.e**(-0.9*x**2 - 0.45*(x-y)**2)
f
```

```python
plot3d(f,(x,-3,3),(y,-3,3))
```

```python
f = 2*x + 5*y + 1
f 
```

```python
solve(f)
```

```python
f = x**2 + y**2 - 1
f 
```

```python
solve(f)
```

```python
f = x*cos(y)
f
```

```python
solve(f)
```
