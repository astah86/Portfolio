# Задача 2

f = a2 * x2 + a1 * x1 + a0

```python
a2, a1, a0, x1, x2 = symbols('a2, a1, a0, x1, x2')
```

```python
MSE = (1/3) * ((1945 * a2 + 560 * a1 + a0 - 4.3)**2 + 
               (1495 * a2 + 340 * a1 + a0 - 4.9)**2 + 
               (1570 * a2 + 343 * a1 + a0 - 5.2)**2)
               
MSE
```

```python
MSEa2 = diff(MSE,a2)
MSEa1 = diff(MSE,a1)
MSEa0 = diff(MSE,a0)
display(MSEa2)
display(MSEa1)
display(MSEa0)
```

```python
eq = linsolve([MSEa2, MSEa1, MSEa0], (a2, a1, a0))
eq
```

```python
f = eq.args[0][0] * x2 + eq.args[0][1] * x1 + eq.args[0][2]

display(f)
MSE.subs({'a2': eq.args[0][0], 'a1': eq.args[0][1], a0: eq.args[0][2]})
```

```python
result = f.subs({'x2': 1520, 'x1': 431})
result
```
